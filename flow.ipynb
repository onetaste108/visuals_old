```python
%load_ext autoreload
%autoreload 2
import img_utils as im
import cv2
import os
import numpy as np
```

```python
init = im.size(im.load("data/in02/01.jpg"), (1024, 1024))
flows = [np.float32(im.size(im.load("data/in02/noise2/"+f), (1024,1024)))/255-0.5 for f in os.listdir("data/in02/noise2/")]
```

```python
def warp_flow(img, flow):
    h, w = flow.shape[:2]
    flow = -flow
    flow[:,:,0] += np.arange(w)
    flow[:,:,1] += np.arange(h)[:,np.newaxis]
    print(flow.shape)
    res = cv2.remap(img, flow, None, cv2.INTER_LANCZOS4, cv2.BORDER_WRAP)
    return res
```

```python
out = init
for f in flows:
    for i in range(10):
        out = warp_flow(out, np.float32(f[:,:,:2])*10)
    im.save_frame(out, "data/out02/flow07")
```
